### 2.1. Value of 0 for "beds", "bathrooms", or "price"

```python
zero_values = db.Customers.aggregate([

# Here we filter documents where beds, bathroom and price contains a value of 0
    {"$match": {"$or": [{"beds": 0}, {"bathrooms": 0}, {"price": 0}]}},

# Then we group the selected documents into one group, that is represented by the _id value of 0, then the sum operator calculates the total count of documents in this group, that match our conditions.   
    {"$group": {"_id": 0, "count": {"$sum": 1}}}
    
])
pprint(list(zero_values))
```

```python
# Now we can delete the records, where the condition we defined above happens. 
db.Customers.delete_many({"$or": [{"beds": 0}, {"bathrooms": 0}, {"price": 0}]})
```

### 2.1. Maximum_nights

The goal is to find if there are extreme values, in this case we would like to analize the highest value and also the most common one.

```python
type(db.Customers.find_one().get("maximum_nights"))
```

```python
value_counts = db.Customers.aggregate([
# We need to convert the maximum numbers of nights to a integer since there are records that are not integers.
    {"$project": {"maximum_nights_to_int": {"$toInt": "$maximum_nights"}}},

# Here we group by the field we want to analyze, this will be a process repeated a few times on this exercise. We count each time a value appears in the collection.
    {"$group": {"_id": "$maximum_nights_to_int", "count": {"$sum": 1}}},

# We sort the values in descending order, because it will become easier to see the high values.
    {"$sort": {"_id": -1}},

# The goal here would be to increave readabillity
    {"$project": {"_id":0, "Maximum Number of Nights": "$_id", "Count": "$count"}}
])

pprint(list(value_counts))
```

We noticed that there are values really high, resulting in ``maximum nights`` of over 3 years. We believe the values are high, but since we have 3137 records with a value of 1125, we would be deleting a great amount of data. We decided that the best option would be to delete values above the 1125 days threshold.  

We can't forget that ``maximum nights`` is not an integer, so we need to use ``parseInt`` to convert it to an integer. 

```python
result = db.Customers.find({
# Find where the maximum nights field is greater than 1125
    "$where": "parseInt(this.maximum_nights) > 1125"
}, 
# Just show the maximum nights field
{"_id": 0, "maximum_nights": 1})


result = db.Customers.delete_many({
    "$where": "parseInt(this.maximum_nights) > 1125"
})

pprint(f"{result.deleted_count} documents were deleted from database.")
```

### 2.4. Summary

We want to analyze if this field is missing. We believe that summary is important to have values, but we also can be flexible, so our decision will be done together with description, when we get there.

But we also want to analyze if there are duplicated records.

```python
missing_summary = db.Customers.aggregate([

# Here we find documents where summary does not exist or is null or contians one of the "values" on the list we defined bellow.
    {"$match": {"$or": [{"summary": {"$exists": False}}, {"summary": {"$in": [".", " ", "  ", "?",""]}}]}},

    {"$project": {"summary": 1}}
])

pprint(list(missing_summary))
```

We can see there are a few missing values in summary, we will keep them for now, and then on details pair this discovery to take actions.  

Now let's check for duplicated values, taking id into account.

```python
duplicates_summary = db.Customers.aggregate([

# We group the values by summary and count how many times each summary appears in the collection.
    {"$group": {"_id": "$summary", "count": {"$sum": 1}}},

# We only want to see duplicated values, so we ask for a count greater than 1.
    {"$match": {"count": {"$gt": 1}}},

# Sum the number of duplicated values.
    {"$group": {"_id": None, "Number of duplicated values": {"$sum": 1}}},

    {"$project": {"_id": 0, "Number of duplicated values": 1}}
])

pprint(list(duplicates_summary))
```

We discovered that we have 34 duplicates. Let's analyze our options. 

We think that some hosts might have several houses that are similar to each other, for example, a host owns two apartments on the same building, the odds that these summaries are the same is very high.  

Another perspective would be that different hosts that have similar houses, will give the same summary as it's neighboor if he also rents, for example, if two neighboor host rent, the apartments are the same, so to chances one copies the other one, are also high.  

Based on both these options we decided to not remove the records, since there were not many.

```python
# For this question, we were asked for two different things: "What is the average duration of stay (in nights) per type of property per city?" and "For
# each property type, return the city with the highest and lowest average value." Let's start with the first query, where we want to find the average
# duration of stay for each property in each city. Here, we will use $match to keep only records where the "maximum_nights" field is present, but also
# where "city" is not just an empty space. Next, we will group by property and city, and we will calculate the average duration of stay. Since the field
# "maximum_nights" is deemed as a string, we will have to convert these records to integers to be able to calculate the mean. Then, we will select only
# the fields we are interested in (and round the average duration of stay to 2 decimal places) and sort by city and, within each one, by property type,
# in order to get a cleaner visualization of the results.
question7_1 = list(db.Customers.aggregate([
    {"$match": {"$expr": {"$ne": [{"$toInt": "$maximum_nights"}, None]},
                "address.market": {"$ne": ""}}},
    {"$group": {"_id": {"property_type": "$property_type", "city": "$address.market"},
                "avg_stay": {"$avg": {"$toInt": "$maximum_nights"}}}},
    {"$project": {"Property Type": "$_id.property_type",
                  "City": "$_id.city",
                  "Average Stay": {"$round": ["$avg_stay", 2]},
                  "_id": 0}},
    {"$sort": {"City": 1, "Property Type": 1}}
]))

pprint(question7_1)
```

In order to solve this question, we'll need to go through 2 phases. For ease of reading purposes, we'll solve this question in 3 separate cells.
Since we dispose of different review scores for different metrics, we believe it's best to merge them into an embedded field. As such, on the first cell we will create a function to help us achieve this more efficiently.
On the second cell, we will dedicate ourselves to create a function that will calculate the average for the review score

Let's start by merging the review score fields.s.

```python
#UpdateOne is a class provided by the PyMongo library which allows us to perform an update operation to a single document in a collection, more efficiently.
from pymongo import UpdateOne

#For ease of solving, let's start by creating a function to merge the fields related with review scores.
def merge_review_scores():
    # Define the update operations to merge review scores fields
    update_operations = [
        UpdateOne(
            {},
            [
                {"$set": {"review_scores": {"$mergeObjects": [
                    {"checkin": "$review_scores_checkin"},
                    {"cleanliness": "$review_scores_cleanliness"},
                    {"communication": "$review_scores_communication"},
                    {"location": "$review_scores_location"},
                    {"value": "$review_scores_value"}
                ]}}},
                {"$unset": [
                    "review_scores_checkin", "review_scores_cleanliness",
                    "review_scores_communication", "review_scores_location",
                    "review_scores_rating", "review_scores_value"
                ]}
            ]
        )
    ]

    # Execute the bulk update operation
    db.listingsAndReviews_new.bulk_write(update_operations)
```

Now, we'll create a function to compute the average review scores.

```python
#Now, we'll create a function to calculate for the inserted property the average review score.
def calculate_average_review_score(property_id):    
# Define the aggregation pipeline to compute the average review score
    pipeline = [
        {"$match": {"_id": property_id}},
        {"$project": {"review_scores_array": {"$objectToArray": "$review_scores"}}},
        {"$unwind": "$review_scores_array"},
        {"$group": {
            "_id": "$_id",
            "avg_review_score": {"$avg": "$review_scores_array.v"}
        }},
        {"$project": {"_id": 0, "avg_review_score": {"$round": ["$avg_review_score", 2]}}}
    ]

    # Execute the aggregation pipeline
    result = list(db.listingsAndReviews_new.aggregate(pipeline))
    return result
```

Next, in order to identify each property type (and the respective number of those properties) for each top 10 city we must define a pipeline which aggregates the number of properties for each property type in intended cities. It must also calculate the total count of properties for each city,  and sort the cities by the total number of properties in descending order (as it was done on the previous exercise), so that the output retrieves only the results related to the top 10 cities.

```python
pipeline_property_type = [
    {"$group": {
        "_id": {"city": "$address.market", "property_type": "$property_type"},
        "count": {"$sum": 1}
    }},
    {"$group": {
        "_id": "$_id.city",
        "property_types": {
            "$push": {
                "type": "$_id.property_type",
                "count": "$count"
            }
        },
        "total_properties": {"$sum": "$count"}
    }},
    {"$sort": {"total_properties": -1}},
    {"$limit": 10}
]

top10_property_type = list(db.listingsAndReviews_new.aggregate(pipeline_property_type))

pprint(top10_property_type)
```

Below is a script designed to identify the top 10 cities with the most listed properties. For each of these cities, it determines the top 3 properties based on their average rating. The script then compiles this information into a list, including details such as property ID, name, rating, number of reviews, superhost status, property type, and price. This information can be valuable for users seeking examples of available properties in these popular destinations.

```python
# Define a function to calculate the average rating for a property
def calculate_average_rating(reviews):
    total_rating = 0
    total_reviews = 0
    for review in reviews:
        if 'review_scores_rating' in review:
            total_rating += review['review_scores_rating']
            total_reviews += 1
    if total_reviews == 0:
        return 0
    return total_rating / total_reviews

# Find the top 10 cities
top10_cities = db.listingsAndReviews_new.aggregate([
    {"$group": {"_id": "$address.market", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$limit": 10}
])

# Create a list to store the results
result = []

# Iterate over the top 10 cities
for city in top10_cities:
    city_name = city['_id']
    # Find all properties in the city
    properties = db.listingsAndReviews_new.find({"address.market": city_name})
    # Create a list to store property information
    city_properties = []
    # Iterate over properties to find top 3 based on average rating
    for property in properties:
        property_id = property['_id']
        property_name = property['name']
        number_of_reviews = property['number_of_reviews']
        price = property['price']
        host_is_superhost = property['host_is_superhost']
        property_type = property['property_type']
        # Calculate average rating for the property
        average_rating = calculate_average_rating(property['reviews'])
        # Add property information to the list
        city_properties.append({
            "Property ID": property_id,
            "Name": property_name,
            "Rating": average_rating,
            "Number of reviews": number_of_reviews,
            "Superhost": host_is_superhost,
            "Type of property": property_type,
            "Price": price
        })
    # Sort properties by rating in descending order
    city_properties.sort(key=lambda x: x["Rating"], reverse=True)
    # Take top 3 properties
    top3_properties = city_properties[:3]
    # Add city and top 3 properties to the result list
    result.append({"City": city_name, "Properties": top3_properties})

# Print the result
pprint(result)
```
